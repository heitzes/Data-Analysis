```python
import pandas as pd
```

## excel을 읽는다

```python
df = pd.read_excel('panda.xlsx')
df.dropna()
# NaN 제거하기위해 dropna() 함수를 사용한다.
```

```python
df.columns
```

## column 이름 바꾸기

```python
df.rename(columns={'Unnamed: 1':'기관명', 'Unnamed: 2':'대표자','Unnamed: 3':'도','Unnamed: 4':'시','Unnamed: 5':'주소',
                  'Unnamed: 6':'전화번호','Unnamed: 7':'신청 유효기간','Unnamed: 8':'기관유형','Unnamed: 9':'실습비',
                  'Unnamed: 10':'실습지도자 인원','Unnamed: 11':'실습지도자 명단'}, inplace=True)
```

```python
df.describe()
## describe 함수는 basic statics (count, mean, max 등등) 을 제공해준다.
```

## '시' column을 기준으로 sorting

```python
df.sort_values(['시'])
```

## '시'에 해당되는 값이 용인시, 성남시인 case(aa,bb)만 모아서 yongin, seongnam에 저장

> aa는 현재의 df에서(전체 dataset(원본))에서 '시'에 해당되는 값이 용인시면 True를 return한다.

> df에서 aa의 return값이 true인 경우(용인시)를 모아서 이를 새로운 dataframe인 yongin에 저장한다.

```python
aa = df['시']=='용인시'
bb = df['시']=='성남시'
yongin = df[aa]
seongnam=df[bb]
```

## 이제, 용인시만 모아둔 yongin에서 주소에 기흥구가 있는 case만 필터링 하고싶다.

> 위의 aa = df['시']=='용인시' 와 달리, 주소 자체가 기흥구 인 것이 아니라 주소에 기흥구라는 단어가 포함된 경우이다.

## 특정 문자열을 기준으로 row를 필터링 할 땐 str.contains('~~')을 사용한다
 
> 문자열이 포함되어 있는 row 조회
* yongin[yongin['주소'].str.contains('기흥구')]

> 문자열이 포함되지 않는 row 조회
* yongin[yongin['주소'].str.contains('기흥구') == False]

```python
yongin[yongin['주소'].str.contains('기흥구')]
```

## 새로운 excel 파일로 저장할때는 to_excel 함수를 이용한다

```python
ki = yongin[yongin['주소'].str.contains('기흥구')]
ki.to_excel('기흥구.xlsx')
```

```python
su = yongin[yongin['주소'].str.contains('수지구')]
su.to_excel('수지구.xlsx')
```

```python
bun = seongnam[seongnam['주소'].str.contains('분당구')]
bun.to_excel('분당구.xlsx')
```
